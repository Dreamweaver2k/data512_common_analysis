# Introduction
This file is for estimating smoke impact on Pullman, WA every year from the distance of fires, number of fires, and acres burned by fires. Every fire is given an impact score calculated by $\texttt{impact} = e^{-d + a}*a$ where d is the distance of the fire and a is the acres burned. Distance and acres are normalized so that they are between 0 and 1. The impact score for each fire is summed to calculate the total impact of smoke in a given year. I chose this formula because I wanted a formula that decreased exponentially with distance increasing but increased exponentially with area increase. This is entirely from my own brain.

The prediction for future smoke impact is calculated using two methods. First, a simple linear regression which considers the previous 8 years of smoke impact to predict the next year. The second is a SARIMA model. These are both compared in visualizations below.

### Output 
* 'data512_common_analysis/data/smoke_impact_timeseries.csv'
    > A file containing the estimated smoke impact and cumulative acres burned for each year between 1964 and 2020 within 1250 miles of the target city.

```python
import geojson
import numpy as np
import pandas as pd
import json
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# Utility functions for calculating smoke effect and normalizing data.
def smoke_effect(distance, acres):
  return np.exp(-distance + acres)*acres

def normalize_list(input_list):
  min_val = min(input_list)
  max_val = max(input_list)
  
  # Apply min-max scaling to each element of the list
  normalized_list = [(x - min_val) / (max_val - min_val) for x in input_list]
  
  return normalized_list
```

```python
# Load fire data.
f = '../data/fires.json'
with open(f) as f:
    gj = geojson.load(f)
```

```python
# Normalize and calculate smoke impact.
df = pd.DataFrame(gj)
df['d_norm'] = normalize_list(df['Distance'].tolist())
df['a_nrom'] = normalize_list(df['GIS_Acres'].tolist())

df['smoke_impact'] = df.apply(lambda row: smoke_effect(row['d_norm'], row['a_nrom']), axis=1)
df.head()
```

```python
# Find nearby fires and group by year. Save results.
df_adjacent = df[df['Distance'] < 1250]
grouped_df = df_adjacent.groupby('Fire_Year')[['smoke_impact', 'GIS_Acres']].sum().reset_index()
grouped_df.head()
grouped_df.to_csv('../data/smoke_impact_timeseries.csv')
```

```python
# Load AQI data and integrate it to the dataframe.
f_aqi = '../data/aqi_10_worst_avg_2002-2020.json'
with open(f_aqi, 'r') as json_file:
    aqi_data = json.load(json_file)
    aqi_data = {int(key): value for key, value in aqi_data.items()}
aqi_series = pd.Series(aqi_data, name='aqi')

df_aqi = grouped_df.copy()
df_aqi['aqi'] = grouped_df['Fire_Year'].map(aqi_series)
df_aqi.head()
pd.concat([df, aqi_series], axis=1)

# Calculate the correlation
subset = df_aqi[df_aqi['Fire_Year'] > 2002] # AQI data in pullman not good until 2002
correlation = df_aqi['smoke_impact'].corr(df_aqi['aqi'])

print(f"Correlation between smoke impact and aqi: {correlation}")
```

```python
# Create a simple linear model looking at the previous k years of smoke impact and predicting the next year.
k = 8
X = []
y = []
observations = df_aqi['smoke_impact'].tolist()
labels = df_aqi['smoke_impact'].tolist()

for i in range(k, len(observations)):
    x_entry = []
    for j in range(i-k,i):
        x_entry.append(observations[j])
    X.append(x_entry)
    y.append(labels[i])

X = np.array(X)
y = np.array(y)

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
```

```python
# Predict and plot for the next 30 years.
predictions = []
lower_bound = []
upper_bound = []
X_future = X.copy()
X_future = [1] + X[-1,2:].tolist() + [y_pred[-1]]
for i in range(29):
    X_future = [1] + X_future[2:] + [y_pred[-1]]
    y_pred = model.predict(X_future)
    confidence_interval = model.get_prediction(X_future).conf_int(alpha=0.05)

    l, u = confidence_interval[0]
    lower_bound.append(l)
    upper_bound.append(u)

    predictions.append(y_pred[-1])

plt.plot([x for x in range(2021, 2021+29)], predictions, label='Observed', color='blue')
plt.fill_between([x for x in range(2021, 2021+29)], lower_bound, upper_bound, color='blue', alpha=0.1)

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Linear Model Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train a SARIMA model to forecast smoke index from historical data.
p, d, q = 5, 1, 5 # Non-seasonal orders
P, D, Q, s = 5, 1, 5, 10  # Seasonal orders

sarima_model = SARIMAX(df_aqi['smoke_impact'], order=(p,d,q), seasonal_order=(P, D, Q, s), method='nm')
sarima_results = sarima_model.fit()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Set the number of forecasted years
n = 29  # 2021 - 2049

# Get SARIMA forecast
forecast = sarima_results.get_forecast(steps=n)
a = np.array([x for x in range(2021, 2021 + n)])

plt.figure(figsize=(12, 6))

# Plot observed data using Seaborn
sns.lineplot(data=df_aqi, x='Fire_Year', y='smoke_impact', label='Observed', color='blue', linewidth=2)

# Plot SARIMA forecast using Seaborn
sns.lineplot(x=a, y=list(forecast.predicted_mean), label='SARIMA', color='red', linewidth=2)
plt.fill_between(a, forecast.conf_int().iloc[:, 0], forecast.conf_int().iloc[:, 1], color='red', alpha=0.3)

# Plot Linear Model forecast using Seaborn
sns.lineplot(x=a, y=predictions, label='Linear Model', color='orange', linewidth=2)
plt.fill_between(a, lower_bound, upper_bound, color='orange', alpha=0.1)

# Customize labels and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Impact Score', fontsize=14)
plt.title('Smoke Impact Forecast', fontsize=16)

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
